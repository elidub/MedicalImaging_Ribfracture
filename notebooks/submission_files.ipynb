```python
import matplotlib.pyplot as plt
import os
import sys
import numpy as np

sys.path.insert(1, sys.path[0] + '/..')
from src.data.utils import read_image
```

```python
submission, _ = read_image('../logs/submissions/version_1/RibFrac424_pred.nii.gz')
label, _ = read_image('../data_dev/raw/val/labels/RibFrac424-label.nii.gz')
image, _ = read_image('../data_dev/raw/val/images/RibFrac424-image.nii.gz')

for z in range(label.shape[2]):
    if np.sum(label[:, :, z]) > 0:
        print(z)
        break
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image[:, :, z], cmap='gray')
ax[1].imshow(label[:, :, z], cmap='gray')
ax[2].imshow(submission[:, :, z], cmap='gray')
ax[0].set_title('Image')
ax[1].set_title('Label')
ax[2].set_title('Submission')
plt.show()
```
