```python
ratios = ['WH', 'WD', 'HD']
```

```python
def calc_ratio(dims, mode):
    if mode == 'WH':
        return dims[0] / dims[1]
    if mode == 'WD':
        return dims[0] / dims[2]
    if mode == 'HD':
        return dims[1] / dims[2]


def avg_ratio_pi(data, im_id, mode='WH'):
    ''' Average ratio per image, on a set mode. '''
    ratios = []
    for frac in data[im_id]:
        dims = frac[3:]
        ratio = calc_ratio(dims, mode)
        ratios.append(ratio)
       
    if len(ratios) == 1:
        return s.mean(ratios), 0
    elif len(ratios) > 0:
        return s.mean(ratios), s.stdev(ratios)
    else:
        return -1, -1
    
def avg_ratio_pm(data, mode='WH'):
    ''' Average ratio pet mode, per image. '''
    means = []
    stdevs = []
    
    for image in data:
        mean, stdev = avg_ratio_pi(data, image, mode)
        means.append(mean)
        stdevs.append(stdev)
    
    means = [i for i in means if i >= 0]
    stdevs = [j for j in stdevs if j >= 0]
    
    return means, stdevs

    
avg_ratio_pm(data)
```
